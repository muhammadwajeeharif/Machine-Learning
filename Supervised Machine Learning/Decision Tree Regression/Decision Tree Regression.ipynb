```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Poly_dataSet.csv')
X = dataset.iloc[:, 0:1].values
y = dataset.iloc[:, 1].values
```

```python
# Fitting DT Regression to the dataset
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor(random_state =0)
dt_reg.fit(X, y)
```

```python
# Visualising the DT Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, dt_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the DT Regression results
# With high Resolution
x_grid=np.arange(min(X),max(X), 0.1)
x_grid =x_grid.reshape(len(x_grid),1)
plt.scatter(X, y, color = 'red')
plt.plot(x_grid, dt_reg.predict(x_grid), color = 'blue')
plt.title('Truth or Bluff (DT Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
